```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

```

```python

# Example high-dimensional vectors
vectors = np.array([
    [0.12, -0.45, 0.89, -0.23],  # "king"
    [0.11, -0.44, 0.88, -0.22],  # "queen"
    [0.10, -0.50, 0.85, -0.20],  # "man"
    [0.09, -0.47, 0.84, -0.21],  # "woman"
    [-0.5, 0.3, -0.8, 0.1]      # "apple" (unrelated word)
])
labels = ["king", "queen", "man", "woman", "apple"]

```

```python
# Reduce to 3D instead of 2D
pca = PCA(n_components=3)
reduced_vectors = pca.fit_transform(vectors)
```

```python
# Plotting in 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
for vector, label in zip(reduced_vectors, labels):
    ax.scatter(vector[0], vector[1], vector[2])
    ax.text(vector[0], vector[1], vector[2], label, fontsize=12)

ax.set_title("3D Visualization of Word Vectors")
ax.set_xlabel("PCA Component 1")
ax.set_ylabel("PCA Component 2")
ax.set_zlabel("PCA Component 3")
plt.show()
```
