- Convert BGR to RGB (OpenCV loads images in BGR format)

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image_bgr = cv2.imread('/content/drive/MyDrive/Colab Notebooks/ComputerVision/cv-basics-image-1.jpg')
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)


# Subplot 1: RGB Image
plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title("Image - RGB")
plt.axis("off")

# Subplot 2: BGR Image
plt.subplot(1, 3, 2)
plt.imshow(image_bgr)
plt.title("CV Image - BGR")
plt.axis("off")

# Subplot 3: Gray scale Image
plt.subplot(1, 3, 3)
plt.imshow(gray_image, cmap='gray')
plt.title("Gray scale Image")
plt.axis("off")

plt.show()
```

# Edge detection

```python
edges_1 = cv2.Canny(gray_image, 80, 200)
edges_2 = cv2.Canny(gray_image, 20, 200)
edges_3 = cv2.Canny(gray_image, 130, 200)


plt.subplot(1, 3, 1)
plt.imshow(edges_1, cmap='Blues')
plt.title("TL-80 TH-200")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(edges_2, cmap='Blues')
plt.title("TL-20 TH-200")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(edges_3, cmap='Blues')
plt.title("TL-130 TH-200")
plt.axis("off")


plt.show()
```
